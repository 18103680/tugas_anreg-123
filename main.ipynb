## maka pada ini tidak multikolineritas pada data tersebut 

```python
# Misalkan df adalah DataFrame yang sudah didefinisikan sebelumnya
# Misalkan 'x1', 'x2', dan 'x3' adalah nama kolom di DataFrame df
# Pastikan kolom 'y' juga ada di df

# Ambil kolom 'x1', 'x2', dan 'x3' dari DataFrame dan tambahkan kolom bias
X = np.column_stack((np.ones(df.shape[0]), df[['x1', 'x2', 'x3']]))

# Ambil nilai-nilai untuk Y
Y = np.array(df[['y']])

# Menghitung (X^T * X)
X_transpose = X.T
XTX = X_transpose @ X

# Menghitung invers dari (X^T * X)
XTX_inv = np.linalg.inv(XTX)

# Menghitung (X^T * Y)
XTY = X_transpose @ Y

# Menghitung hasil akhir (koefisien regresi)
result = XTX_inv @ XTY

# Menampilkan hasil
print("Hasil:")
print(result)
```
