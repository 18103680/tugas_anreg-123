```python
import pandas as pd
from scipy import stats

# Misalkan kita punya DataFrame df
# df = pd.DataFrame(...)

# Pilih kolom yang ingin diuji (misal 'x1')
data = df['x2']

# Lakukan uji normalitas menggunakan Shapiro-Wilk
stat, p_value = stats.shapiro(data)

print(f'Statistik Uji: {stat}, p-value: {p_value}')

# Interpretasi hasil
alpha = 0.05
if p_value > alpha:
    print('Data berdistribusi normal (tidak dapat menolak H0)')
else:
    print('Data tidak berdistribusi normal (menolak H0)')
```

## maka pada ini tidak multikolineritas pada data tersebut 

```python
# Misalkan df adalah DataFrame yang sudah didefinisikan sebelumnya
# Misalkan 'x1', 'x2', dan 'x3' adalah nama kolom di DataFrame df
# Pastikan kolom 'y' juga ada di df

# Ambil kolom 'x1', 'x2', dan 'x3' dari DataFrame dan tambahkan kolom bias
X = np.column_stack((np.ones(df.shape[0]), df[['x1', 'x2', 'x3']]))

# Ambil nilai-nilai untuk Y
Y = np.array(df[['y']])

# Menghitung (X^T * X)
X_transpose = X.T
XTX = X_transpose @ X

# Menghitung invers dari (X^T * X)
XTX_inv = np.linalg.inv(XTX)

# Menghitung (X^T * Y)
XTY = X_transpose @ Y

# Menghitung hasil akhir (koefisien regresi)
result = XTX_inv @ XTY

# Menampilkan hasil
print("Hasil:")
print(result)
```
