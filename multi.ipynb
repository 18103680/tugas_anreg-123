```python
import numpy as np
import pandas as pd

# Misalkan df adalah DataFrame yang sudah didefinisikan sebelumnya
# Misalkan 'x1' dan 'x2' adalah nama kolom di DataFrame df
# X = np.array(df[['x1', 'x2']])  # Ambil kolom 'x1' dan 'x2' dari DataFrame
X = np.column_stack((np.ones(df.shape[0]), df[['x1', 'x2']]))
Y = np.array(df[['y']])  # Nilai-nilai untuk Y

# Tampilkan X untuk memverifikasi
print(X)
```

```python
# Menghitung (X^T * X)
X_transpose = X.T
XTX = X_transpose @ X

# Menghitung invers dari (X^T * X)
XTX_inv = np.linalg.inv(XTX)

# Menghitung (X^T * Y)
XTY = X_transpose @ Y

# Menghitung hasil akhir
result = XTX_inv @ XTY

# Menampilkan hasil
print("Hasil:")
print(result)
```

```python

# Membuat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotting data asli
ax.scatter(df['x1'], df['x2'], df['y'], c='b', marker='o', label='Data Asli')

# Membuat grid untuk x1 dan x2
x1_range = np.linspace(df['x1'].min(), df['x1'].max(), 100)
x2_range = np.linspace(df['x2'].min(), df['x2'].max(), 100)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Menghitung nilai y berdasarkan persamaan regresi
y_grid = -34.91789758+ 0.15644203 * x1_grid - 0.12788049 * x2_grid

# Plotting permukaan regresi
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5, rstride=100, cstride=100, color='r', label='Regresi')

# Menambahkan label
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()

# Menampilkan plot
plt.show()
```
